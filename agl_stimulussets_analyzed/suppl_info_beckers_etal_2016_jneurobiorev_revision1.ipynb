Supplementary Analysis 1

The string sets originate from Experiment 3 in: Saffran J, Hauser M, Seibel R, Kapfhamer J, Tsao F, Cushman F (2008) Grammatical pattern learning by human infants and cotton-top tamarin monkeys. Cognition 107:479â€“500. [link](http://www.sciencedirect.com/science/article/pii/S0010027707002697)

In Experiment 3 of this paper, cotton-top tamarins are familiarized with two languages, a P-language, which contains predictive dependencies, and a NP-language, which does not contain such predictive dependencies. Tamarins were tested with grammatical and ungrammatical test strings one day after familiarization with either a P-language or a NP-language. Grammatical test strings conform to the grammar of both languages and ungrammatical test strings violate the grammar of both languages. The idea behind this experiment is to test if the P-language, is more easily learned by the tamarins. If so, tamarins should distinguish between the grammatical and ungrammatical strings better after they have been familiarized with the P-language.

In this particular experiment (but not in all other ones reported in the same paper), the tokens from which the strings constructed are particular sounds, not sound classes. The sounds are the speech syllables "biff", "cav", "dupp", "klor", and "jux" (see their Table 1 [link](http://www.sciencedirect.com/science/article/pii/S0010027707002697#tbl1)), here denoted as {b,c,d,k,j}.

The results show that: "In the P-Language condition, tamarins responded significantly more to the ungrammatical items than the grammatical items [Wilcoxon, z = 3.19, p = .001]. In contrast, they showed no statistically significant pattern of response to the NP language [Wilcoxon, z = 0.4, p = .69]. An ANOVA examining the interaction between items (grammatical and ungrammatical) and conditions (P-language and NP language) was significant [F(1, 33) = 4.56, p < .05]."

The authors conclude (p. 491) that this suggests "that tamarins are able to detect regularities in a simple grammar, written over individual word tokens, when predictive dependencies are present."

However, there is a bias in how often test strings are fully contained within familiarization strings (Table S1): ungrammatical test strings never fully occur in any familiarization string, but the grammatical test strings do, and not in a balanced way. In half the of grammatical test strings (GTS_1 and GTS_3) this occurs three times as often in P-language strings than in NP-language strings. Thus, before they are tested the tamarins have been familiarized more extensively with complete grammatical test string sounds from the P-language than with complete grammatical test string sounds from the NP-language. This bias does not exist in the ungrammatical test strings (UTS), which are never fully present in either language. This provides a non-grammatical explanation for the results.

The string sets originate from:  Wilson B, Slater H, Kikuchi Y, Milne AE, Marslen-Wilson WD, Smith K, Petkov CI (2013) Auditory Artificial Grammar Learning in Macaque and Marmoset Monkeys. J Neurosci 33:18825â€“18835. [link](http://www.jneurosci.org/content/33/48/18825.full)

In this paper, the authors first habituated macaque and marmoset monkeys to strings conforming to grammatical rules, and then presented them with test strings that are either consistent with these rules or that contain illegal violations. The goal is to test if monkeys can learn aspects of the artificial grammar. The macaques did differentiate between grammatical and ungrammatical test string in a way that the authors interpret as "... evidence for a previously unknown level of AGL complexity in Old World monkeys that seems less evident in New World monkeys ...".

The elements that were used to build the strings ({a,c,d,g,f}) stand for short nonsense words (a: "klor", c: "biff", d: "jux", f: "cav", g: "dupp").

The main figure of the results on which their conclusion is based, their Figure 2, is openly accessible:  [link](http://www.jneurosci.org/content/33/48/18825/F2.expansion.html).

The analysis here (Table S3 and Figure S1) shows that grammar consistent test strings (Fam and Novel) share longer  substrings with the habituation strings than grammar violating test strings do, which provides a non-grammatical explanation for the results.

The strings (from their Figure 1 [link](http://www.jneurosci.org/content/33/48/18825/F1.expansion.html)) are here labeled as follows:

Habituation strings (rows):

- Hab

Test strings (columns):

- <font color="#0071c1">Fam: familiar test strings' (i.e. identical to Hab strings)</font>
- <font color="#01b0f1">Novel: grammar consistent test strings</font>
- <font color="#c00000">Viol_bA: grammar violating test strings that begin with an _a_ </font>
- <font color="#fe0000">Viol_nbA: grammar violating test strings that do not begin with an _a_ </font>

They conclude: "We observed a considerable level of cross-species correspondence in the sensitivity of both humans and macaques to the adjacent AG relationships and to the statistical properties of the sequences." ... "The results suggest that humans and macaques are largely comparably sensitive to the adjacent AG relationships and their statistical properties. However, in the presence of multiple cues to grammaticality, the non-adjacent relationships are less salient to the macaques and many of the humans."

Supplementary Analysis 5
---

The string sets originate from:  Chen J, ten Cate C (2015) Zebra finches can use positional and transitional cues to distinguish vocal element strings. _Behavioural Processes_, 117:29â€“34. [link](http://www.sciencedirect.com/science/article/pii/S0376635714001995)

In this paper the authors use Go-Nogo operant tests to examine whether zebra finches can use transitional cues to distinguish artificially constructed strings of natural song elements. From their abstract: "Zebra finches were trained with sequences differing in transitional and positional information and next tested with novel strings sharing positional and transitional similarities with the training strings. The results show that they can attend to both transitional and positional cues and that their sequential coding strategies can be biased toward transitional cues depending on the learning context."


Their Experiment 1
---

<a id="TS5"></a>The strings of Experiment 1 are given in their Table 1 [link](http://www.sciencedirect.com/science/article/pii/S0376635714001995/#tbl0005) (and printed in Table S5 below), and the results are provided in their Table 4 [link](http://www.sciencedirect.com/science/article/pii/S0376635714001995/#tbl0020). Birds were trained to distinguish the _Go_ sound from the _NoGo_ sound, and were then tested on probes _A1_, _A2_, _B1_, and _B2_. Differential responses to <i>A</i> and <i>B</i> sounds should distinguish between the use of element transitions and positions. See their Table 1 and text in section 2.3.1 for the rationale behind this design.

However, there is a bias in the length of the substrings shared between test and probe strings (see Tables S6 and Figure S4 below).

The behavioural results (Go: 0.87, NoGo:0.10, A1: 0.70, A2: 0.67, B1: 0.57, B2: 0.49; see their Table 4 [link](http://www.sciencedirect.com/science/article/pii/S0376635714001995/#tbl0020)) match this pattern very well and provides a non-grammatical alternative explanation for the results.

The string sets originate from the study of  Abe K, Watanabe D (2011) Songbirds possess the spontaneous ability to discriminate syntactic rules. Nat Neurosci 14:1067â€“1074. [<a href="http://www.nature.com/neuro/journal/v14/n8/full/nn.2869.html">link</a>].

Abe and Watanabe (2011) tested whether or not Bengalese finches spontaneously learn a context-free grammar from the examples they were presented with (NEF and EF stimuli, see strings below). They did this by measuring the birds' responses to test stimuli, some of which conform to the grammar (COR and CES stimuli), and some of which do not (AFP, AFC and RAND stimuli). The results provided in their Figure 3d show that grammar-conforming COR and CES stimuli are not treated differently from training stimuli, but grammar-violationg AFP, AFC, and RAND stimuli are.

However, there is a bias in the maximum crosscorrelation between test and familiarization strings so that the results can be explained on the bases of acoustic similarity alone (Fig. S9).

A crosscorrelation between two strings is calculated by fully sliding one across the other in a stepwise fashion and counting at every step how many elements correspond between them. From the resulting counts, the maximum is taken. For example, the crosscorrelation between the strings _A1C1F1_ and _A1C2F1_ is: [0,0,2,0,0]. Its maximum is 2.

The strings were not published in Abe and Watanabe (2011), but they were graciously provided to us and we analysed them in Beckers et al (2012). For completeness, we also print them here:

```python
strings = get_examplestringdata('abewatanabe_natneurosci_2011')
strings.print_strings(comparison='Fig3d')
```

The strings sets originate from: Chen J, Rossum D van, Cate C ten (2015) Artificial grammar learning in zebra finches and human adults: XYX versus XXY. Anim Cogn 18:151â€“164.  [link](http://link.springer.com/article/10.1007/s10071-014-0786-4/fulltext.html)
    
In a series of tests, the authors present zebra finches with training and test strings with XYX and XXY structures, in which X and Y were zebra finch song elements (see their Figure 1 [link](http://link.springer.com/article/10.1007/s10071-014-0786-4/fulltext.html#Fig1)). Zebra finches discriminated the training stimuli, and some birds discriminated novel test stimuli when these were composed of familiar element types. None of the birds generalized the discrimination to novel element types. The authors conclude that zebra finches show evidence of simple rule abstraction related to positional learning, but found no evidence for a more abstract rule generalization.

However, there is a bias in acoustic similarity between training and test strings, based on the level of maximum crosscorrelation, that can explain or may have contributed to the results, without the use of rule abstraction.

The analyses relevant to the discussion here are those of their Experiment 1b, because the results of this experiment appear the provide the basis for the authors' claim of rule abstraction. Experiment 1a is addressed below for completeness, and to show that our alternative explanation is consistent with those results as well.
